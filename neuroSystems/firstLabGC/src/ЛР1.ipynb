### 1. Создание схемы нейронной сети согласно варианту таблицы

Рассмотрим двухслойную нейронную сеть следующей структуры:
- **Входной слой:** 3 входа ($x_1, x_2, x_3$).
- **Скрытый слой (первый):** 7 нейронов, функция активации — сигмоида: $P(S) = \frac{1}{1 + e^{-S}}$.
- **Выходной слой (второй):** 2 нейрона, функция активации — пороговая: $Y(S) = \begin{cases} 0, & S < 0,\\ 1, & S \ge 0 \end{cases}$

### 2. Создание математической модели нейронной сети согласно варианту

### 1. Первый (скрытый) слой

В этом слое каждый нейрон $j$ вычисляет взвешенную сумму входов $S_j^{(1)}$, а затем применяет к ней сигмоидную функцию активации для получения выхода $P_j$.

**Развернутая запись для каждого нейрона:**

**Первый нейрон скрытого слоя ($j=1$):**
$$ S_1^{(1)} = x_1 \cdot w_{11}^{(1)} + x_2 \cdot w_{12}^{(1)} + x_3 \cdot w_{13}^{(1)} $$
$$ P_1 = \frac{1}{1 + e^{-S_1^{(1)}}} $$

**Второй нейрон скрытого слоя ($j=2$):**
$$ S_2^{(1)} = x_1 \cdot w_{21}^{(1)} + x_2 \cdot w_{22}^{(1)} + x_3 \cdot w_{23}^{(1)} $$
$$ P_2 = \frac{1}{1 + e^{-S_2^{(1)}}} $$

**Третий нейрон скрытого слоя ($j=3$):**
$$ S_3^{(1)} = x_1 \cdot w_{31}^{(1)} + x_2 \cdot w_{32}^{(1)} + x_3 \cdot w_{33}^{(1)} $$
$$ P_3 = \frac{1}{1 + e^{-S_3^{(1)}}} $$

**Четвертый нейрон скрытого слоя ($j=4$):**
$$ S_4^{(1)} = x_1 \cdot w_{41}^{(1)} + x_2 \cdot w_{42}^{(1)} + x_3 \cdot w_{43}^{(1)} $$
$$ P_4 = \frac{1}{1 + e^{-S_4^{(1)}}} $$

**Пятый нейрон скрытого слоя ($j=5$):**
$$ S_5^{(1)} = x_1 \cdot w_{51}^{(1)} + x_2 \cdot w_{52}^{(1)} + x_3 \cdot w_{53}^{(1)} $$
$$ P_5 = \frac{1}{1 + e^{-S_5^{(1)}}} $$

**Шестой нейрон скрытого слоя ($j=6$):**
$$ S_6^{(1)} = x_1 \cdot w_{61}^{(1)} + x_2 \cdot w_{62}^{(1)} + x_3 \cdot w_{63}^{(1)} $$
$$ P_6 = \frac{1}{1 + e^{-S_6^{(1)}}} $$

**Седьмой нейрон скрытого слоя ($j=7$):**
$$ S_4^{(1)} = x_1 \cdot w_{71}^{(1)} + x_2 \cdot w_{72}^{(1)} + x_3 \cdot w_{73}^{(1)} $$
$$ P_7 = \frac{1}{1 + e^{-S_7^{(1)}}} $$

---

**Обобщенная формула для первого слоя:**

Для любого нейрона $j$ ($j = 1 \dots 7$) скрытого слоя:

$$ S_j^{(1)} = \sum_{i=1}^{3} x_i \cdot w_{ji}^{(1)} $$

$$ P_j = \frac{1}{1 + e^{-S_j^{(1)}}} $$

где:
- $i$ — индекс входа ($i = 1 \dots 3$);
- $j$ — индекс нейрона в скрытом слое ($j = 1 \dots 7$);
- $w_{ji}^{(1)}$ — вес связи, идущей от $i$-го входа к $j$-му нейрону скрытого слоя.

---

### 2. Второй (выходной) слой

В этом слое каждый нейрон $q$ вычисляет взвешенную сумму выходов скрытого слоя $S_q^{(2)}$, а затем применяет пороговую функцию активации.

**Развернутая запись для каждого выхода:**

**Первый выходной нейрон ($q=1$):**
$$ S_1^{(2)} = P_1 \cdot w_{11}^{(2)} + P_2 \cdot w_{12}^{(2)} + P_3 \cdot w_{13}^{(2)} + P_4 \cdot w_{14}^{(2)} + P_5 \cdot w_{15}^{(2)} + P_6 \cdot w_{16}^{(2)} + P_7 \cdot w_{17}^{(2)} $$
$$Y_1 = \begin{cases} 0, & S_1^{(2)} < 0,\\ 1, & S_1^{(2)} \ge 0 \end{cases}$$

**Второй выходной нейрон ($q=2$):**
$$ S_2^{(2)} = P_1 \cdot w_{21}^{(2)} + P_2 \cdot w_{22}^{(2)} + P_3 \cdot w_{23}^{(2)} + P_4 \cdot w_{24}^{(2)} + P_5 \cdot w_{15}^{(2)} + P_6 \cdot w_{16}^{(2)} + P_7 \cdot w_{17}^{(2)} $$
$$Y_2 = \begin{cases} 0, & S_2^{(2)} < 0,\\ 1, & S_2^{(2)} \ge 0 \end{cases}$$

---

**Обобщенная формула для второго слоя:**

Для любого выхода $q$ ($q = 1 \dots 2$) выходного слоя:

$$ S_q^{(2)} = \sum_{j=1}^{7} P_j \cdot w_{qj}^{(2)} $$

$$Y_q = \begin{cases} 0, & S_q^{(2)} < 0,\\ 1, & S_q^{(2)} \ge 0 \end{cases}$$

где:
- $j$ — индекс нейрона скрытого слоя ($j = 1 \dots 7$);
- $q$ — индекс нейрона выходного слоя ($q = 1 \dots 2$);
- $w_{qj}^{(2)}$ — вес связи, идущей от $j$-го скрытого нейрона к $q$-му выходному.

---

**Итоговая формула всей сети (в общем виде):**

Выход $Y_q$ вычисляется как:

$$Y_q = \begin{cases} 0, & \sum_{j=1}^{7} \left( \frac{1}{1 + e^{-\sum_{i=1}^{3} x_i \cdot w_{ji}^{(1)}}} \right) \cdot w_{qj}^{(2)} < 0,\\ 1, & \sum_{j=1}^{7} \left( \frac{1}{1 + e^{-\sum_{i=1}^{3} x_i \cdot w_{ji}^{(1)}}} \right) \cdot w_{qj}^{(2)} \ge 0 \end{cases}$$
где $q = 1 \dots 2$.

## 3. Моделирование нейронной сети (практическое задание)

В этом разделе мы реализуем на Python модель двухслойной нейронной сети согласно варианту.

**Наша задача:**
1. Создать функцию `Neuron`, которая моделирует прямой проход сигнала через сеть (вход → скрытый слой → выход).
2. Создать функцию `Wes`, которая запускает сеть с тремя различными диапазонами случайных весов:
   - **$[-1; 0]$** — только отрицательные (тормозящие) связи;
   - **$[-1; 1]$** — смешанные веса;
   - **$[0; 1]$** — только положительные (возбуждающие) связи.
3. Подать на вход сети тестовые сигналы (линейный, пороговый, синусоидальный) и построить графики выходов.


```python
# Импорты
import random
import numpy as np
import matplotlib.pyplot as plt

# Диапазоны весов (базовый случай)
a = -1
b = 1
```

```python
def Neuron(WS, WV, Xx, vx, weight_label=""):
    """
    Прямой проход двухслойной сети:
    Xx: (3, T)   входы (3 сигнала на T шагов времени)
    WS: (7, 3)   веса вход -> скрытый (7 скрытых нейрона)
    WV: (2, 7)   веса скрытый -> выход (2 выходных нейрона)
    weight_label: подпись диапазона весов для заголовка
    """
    Y = np.zeros(2)
    S = np.zeros(7)
    P = np.zeros(7)
    s = np.zeros(2)
    y = np.zeros((2, Xx.shape[1]))

    for t in range(0, Xx.shape[1]):
        Y = np.zeros(2)
        S = np.zeros(7)
        P = np.zeros(7)
        s = np.zeros(2)

        # Скрытый слой
        for i in range(0, 7):
            for j in range(0, 3):
                S[i] = Xx[j, t] * WS[i, j] + S[i]
            P[i] = 1 / (1 + np.exp(-S[i]))

        # Выходной слой
        for i in range(0, 2):
            for j in range(0, 7):
                s[i] = P[j] * WV[i, j] + s[i]
                if s[i] < 0:
                  Y[i] = 0
                else:
                  Y[i] = 1
            y[i, t] = Y[i]

    # Визуализация выхода
    plt.figure(figsize=(10, 6))
    plt.grid(alpha=0.3)
    plt.xlabel('x', fontsize=12)
    plt.ylabel('y', fontsize=12)
    plt.title(f'Выходы нейронной сети | {weight_label}', fontsize=13, fontweight='bold')
    plt.plot(vx, y[0], linewidth=2, label='Y1', marker='o', markersize=4)
    plt.plot(vx, y[1], linewidth=2, label='Y2', marker='s', markersize=4)
    plt.legend(fontsize=11)
    plt.tight_layout()
```

```python
def Wes(Xx, vx, a, b):
    # [0, b] - только положительные веса
    WS = np.random.uniform(0, b, (7, 3))
    WV = np.random.uniform(0, b, (2, 7))
    Neuron(WS, WV, Xx, vx, weight_label=f"Диапазон весов: [0, {b}]")

    # [a, 0] - только отрицательные веса
    WS = np.random.uniform(a, 0, (7, 3))
    WV = np.random.uniform(a, 0, (2, 7))
    Neuron(WS, WV, Xx, vx, weight_label=f"Диапазон весов: [{a}, 0]")

    # [a, b] - смешанные веса
    WS = np.random.uniform(a, b, (7, 3))
    WV = np.random.uniform(a, b, (2, 7))
    Neuron(WS, WV, Xx, vx, weight_label=f"Диапазон весов: [{a}, {b}]")
```

### 3.2. Подготовка и визуализация входных данных

Сформируем тестовые сигналы, которые будем подавать на вход сети:
1.  **Линейный сигнал** (y = kx + b)
2.  **Пороговый сигнал** (ступенька)
3.  **Синусоидальный сигнал** (y = sin(x))

Построим графики этих входных сигналов, чтобы видеть, с чем будет работать наша нейронная сеть.

```python
k = random.uniform(-10, 10)

Xx = np.random.uniform(0, 1, (3, 20))
vx = np.linspace(-20, 20, 20)

for i in range(0, 20):
    Xx[0, i] = vx[i] * k + 3
    Xx[1, i] = 0.2 if vx[i] < 0 else 1
    Xx[2, i] = np.sin(vx[i])
```

```python
plt.figure()
plt.title('Линейная функция на вход:')
plt.plot(vx, Xx[0], 'bo-', label='y = kx+3')
plt.legend()

plt.figure()
plt.title('Пороговая функция на вход:')
plt.plot(vx, Xx[1], 'r*-', label='y = 0.2 if x < 0 else 1')
plt.legend()

plt.figure()
plt.title('Синусоидальная функция на вход:')
plt.plot(vx, Xx[2], 'r*-', label='y = sin(x)')
plt.legend()
```
